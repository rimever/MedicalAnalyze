```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv( 'ope_by_hospital_size.csv' ,index_col=0,comment ='#')
data = data.query("一般病院_001 == '総数'")
data.head()
```

```python
data.columns
```

```python
data['全身麻酔（静脈麻酔は除く）（実施件数/施設数）'] = data[ '全身麻酔（静脈麻酔は除く）（実施件数）'] / data['全身麻酔（静脈麻酔は除く）（施設数）']
data['全身麻酔（静脈麻酔は除く）（総数/施設数）'] = 100 * data['全身麻酔（静脈麻酔は除く）（施設数）'] / data['総数']

data.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib

font = {'family': 'IPAexGothic'}
matplotlib.rc('font', **font)

fig, ax = plt.subplots(1, 1, figsize=(12, 4))

data['全身麻酔（静脈麻酔は除く）（実施件数/施設数）'] = data[ '全身麻酔（静脈麻酔は除く）（実施件数）'] / data['全身麻酔（静脈麻酔は除く）（施設数）']

plt.title('病床の規模と全身麻酔の件数')
plt.grid()

x = data['病床の規模_002']
y = data['全身麻酔（静脈麻酔は除く）（実施件数/施設数）']
plt.ylabel('全身麻酔（静脈麻酔は除く）（実施件数/施設数）')
plt.xlabel('病床の規模')

plt.scatter(x,y)

```
