```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv( 'span_in_hospital.csv' ,index_col=0,comment ='#')
data.head(20)
```

```python
temp = data.query("分類 == '総数'")
labels = data.columns[2:]
before_ope = temp.iloc[0,2:]
after_ope = temp.iloc[1,2:]

from pylab import rcParams
rcParams['figure.dpi'] = 150

total = [num1 + num2 for num1,num2 in zip(before_ope,after_ope)]

fig,ax = plt.subplots()
plt.xticks(rotation=90)
plt.title('総数')
ax.bar(labels,total,label='after_ope')
ax.bar(labels,before_ope,label='before_ope')
ax.legend()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from japanmap import *

column_name = '合計'

color_map = {}

temp = data.query("分類 == '総数'").T
temp = temp[2:].sort_values(by='術後')
temp[column_name] = temp['術前'] + temp['術後']

mean = np.mean(temp[column_name])
max_value = np.max(temp[column_name])
min_value = np.min(temp[column_name])

for index, row in temp.iterrows():        
    prefecture = str(index).replace('　', '')
    value = row[column_name]
    if mean < value:
        green_value = int(255 - 255 * (value - mean) / (max_value - mean))
        green = str(hex(green_value))[2:]
        if len(green) == 1:
            green = "0" + green
        
        color_map[prefecture] = "#FF" + green + "00"

    else:
        red_value = int(255 - 255 * (mean - value) / (mean - min_value))
        red = str(hex(red_value))[2:]
        if len(red) == 1:
            red = "0" + red
            
        color_map[prefecture] = "#" + red + "FF00"

plt.subplots(figsize=(10, 10))
plt.imshow(picture(color_map), interpolation='nearest', aspect='auto')
```
